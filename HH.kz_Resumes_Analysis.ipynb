```python
# b. Load the csv to the pandas dataframe. And print out the first 20 rows of data frame
df = pd.read_csv('java.csv')
df.head(20)
```

```python
# c. create the scripts to sort dataframe by the salary, age, experience and title (in alphabetical order)
df_sorted_salary = df.sort_values(by='Salary', ascending=False)
df_sorted_age = df.sort_values(by='Age', ascending=False)
df_sorted_experience = df.sort_values(by=['Experience_years', 'Experience_month'], ascending=False)
df_sorted_title = df.sort_values(by='Title', ascending=True)
```

```python
# d. create the scripts to calculate minimum, maximum and mean average of the salary of all
min_salary = df['Salary'].min()
max_salary = df['Salary'].max()
mean_salary = df['Salary'].mean()
```

```python
# e. create the scripts to calculate minimum, maximum and mean average of the salary of each sex, specialization and citizenship. Results must be the dataframe
df_mean_salary_by_sex = df.groupby(['Sex'])['Salary'].mean()
df_mean_salary_by_specialization = df.groupby(['Specialization'])['Salary'].mean()
df_mean_salary_by_citizenship = df.groupby(['Citizenship'])['Salary'].mean()
df_salary_stats = pd.DataFrame({
    'Min Salary': df.groupby(['Sex', 'Specialization', 'Citizenship'])['Salary'].min(),
    'Max Salary': df.groupby(['Sex', 'Specialization', 'Citizenship'])['Salary'].max(),
    'Mean Salary': df.groupby(['Sex', 'Specialization', 'Citizenship'])['Salary'].mean()
})
```

```python
# f. create the scripts to calculate the count of males,females and the CVs of people from each country
df_count_sex = df.groupby(['Sex'])['Sex'].count()
df_count_citizenship = df.groupby(['Citizenship'])['Citizenship'].count()
```

```python
# g. Create at least 5 plots of dataframes from the previous steps.
plt.figure(figsize=(10, 5))
plt.title('Average Salary by Sex')
plt.ylabel('Salary')
df_mean_salary_by_sex.plot(kind='bar')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.title('Average Salary by Specialization')
plt.ylabel('Salary')
df_mean_salary_by_specialization.plot(kind='bar')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.title('Average Salary by Citizenship')
plt.ylabel('Salary')
df_mean_salary_by_citizenship.plot(kind='bar')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.title('Count of CVs by Sex')
plt.ylabel('Count')
df_count_sex.plot(kind='bar')
plt.show()
```
